```python
import yfinance as yf
import pandas as pd
import numpy as np
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# --- Phase 1: Data Prep ---
df = yf.download('NVDA', period='2y', interval='1d')
if isinstance(df.columns, pd.MultiIndex): df.columns = df.columns.get_level_values(0)

# Create Target (Future Move) and Feature (RSI)
df['Target'] = df['Close'].shift(-1) - df['Close']
delta = df['Close'].diff()
gain = delta.where(delta > 0, 0)
loss = -delta.where(delta < 0, 0)
avg_gain = gain.rolling(14).mean()
avg_loss = loss.rolling(14).mean()
df['RSI'] = 100 - (100 / (1 + (avg_gain/avg_loss)))

# Clean data for the Model
data = df[['RSI', 'Target']].dropna()
X = data[['RSI']].values
y = data['Target'].values

# --- Phase 2: The Analysis ---
model = lm.LinearRegression()
model.fit(X, y)
predictions = model.predict(X)
r2 = r2_score(y, predictions)

# --- Phase 3: Visual Output ---
plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.4, label='Historical Data')
plt.plot(X, predictions, color='red', linewidth=3, label=f'Regression Line (R²={r2:.4f})')
plt.title(f'Full Regression: RSI vs Future Move\nSlope: {model.coef_[0]:.4f} | Intercept: {model.intercept_:.4f}')
plt.xlabel('RSI Value')
plt.ylabel('Next Day Price Change ($)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Analysis Complete. Accuracy (R-Squared): {r2:.4f}")
```
