```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fredapi import Fred
 
# ─── CONFIG ───────────────────────────────────────────────────────────────────
 
TICKER    = 'PG'
INTERVAL  = '1d'
PERIOD    = '10y' if INTERVAL == '1d' else '780d'   # FIX: was '1D' → always 780d
 
MACD_FAST = 12
MACD_SLOW = 27
MACD_SPAN = 9
LOOKBACK  = 10000
 
FRED_KEY  = '0673eba354376e034551ab39974ca711'
fred      = Fred(api_key=FRED_KEY)
 
# ─── DATA ─────────────────────────────────────────────────────────────────────
 
def get_data(ticker=TICKER, lookback=LOOKBACK):
    df = yf.download(ticker, interval=INTERVAL, period=PERIOD, progress=False)
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)
 
    df['Asset_Return'] = (1 + df['Close'].pct_change()).cumprod() - 1
    for c in df.columns:
        df[f'{c}_change'] = df[c].pct_change() * 100
 
    subset = df.iloc[-lookback:].copy()
 
    plt.figure(figsize=(12, 4))
    plt.plot(subset['Close'])
    plt.title(f'{TICKER} Price Over Time')
    plt.tight_layout()
    plt.show()
 
    return subset.dropna()
 
 
def add_interest_rate(df):
    fedfunds = pd.DataFrame(fred.get_series('DFF'), columns=['FedFunds'])
    return df.join(fedfunds, how='left').ffill()   # ffill so no NaN gaps
 
 
def add_inflation(df):
    fedinflation = pd.DataFrame(
        fred.get_series('CPIAUCSL', units='pc1'), columns=['Inflation']
    )
    return df.join(fedinflation, how='left').ffill()
 
 
# ─── INDICATORS ───────────────────────────────────────────────────────────────
 
BB_WINDOW = 20
BB_STD    = 2.0

def add_bollinger_bands(df, window=BB_WINDOW, num_std=BB_STD):
    # ── Compute bands ──────────────────────────────────────────────────────
    df['BB_Mid']   = df['Close'].rolling(window).mean()
    rolling_std    = df['Close'].rolling(window).std()
    df['BB_Upper'] = df['BB_Mid'] + num_std * rolling_std
    df['BB_Lower'] = df['BB_Mid'] - num_std * rolling_std

    # %B = where price is within bands (0=lower, 0.5=middle, 1=upper)
    df['BB_PctB']  = (df['Close'] - df['BB_Lower']) / (df['BB_Upper'] - df['BB_Lower'])

    # Bandwidth = how wide the bands are (low = squeeze, big move coming)
    df['BB_Width'] = (df['BB_Upper'] - df['BB_Lower']) / df['BB_Mid'] * 100

    # ── Chart ──────────────────────────────────────────────────────────────
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6), sharex=True,
                                    gridspec_kw={'height_ratios': [3, 1]})

    ax1.plot(df['Close'],    color='black', linewidth=1.2, label='Price')
    ax1.plot(df['BB_Mid'],   color='blue',  linewidth=1.0, linestyle='--', label=f'SMA {window}')
    ax1.plot(df['BB_Upper'], color='red',   linewidth=0.8, label=f'Upper (+{num_std}σ)')
    ax1.plot(df['BB_Lower'], color='green', linewidth=0.8, label=f'Lower (-{num_std}σ)')
    ax1.fill_between(df.index, df['BB_Upper'], df['BB_Lower'], alpha=0.08, color='blue')
    ax1.set_title(f'{TICKER} Price + Bollinger Bands')
    ax1.legend(fontsize=8)

    ax2.plot(df['BB_PctB'], color='purple', linewidth=0.9)
    ax2.axhline(1.0, color='red',   linestyle='--', linewidth=0.8)
    ax2.axhline(0.5, color='grey',  linestyle=':',  linewidth=0.8)
    ax2.axhline(0.0, color='green', linestyle='--', linewidth=0.8)
    ax2.fill_between(df.index, df['BB_PctB'], 1.0,
                     where=(df['BB_PctB'] >= 1.0), alpha=0.2, color='red')
    ax2.fill_between(df.index, df['BB_PctB'], 0.0,
                     where=(df['BB_PctB'] <= 0.0), alpha=0.2, color='green')
    ax2.set_ylabel('%B')

    plt.tight_layout()
    plt.show()

    return df
 
 
# ─── STRATEGY SIGNALS ─────────────────────────────────────────────────────────
 
def economic_strategy(df):
    """
    Macro regime filter:
      Inflation > FedFunds → real rates negative → accommodative → LONG (+1)
      FedFunds > Inflation → real rates positive → restrictive   → SHORT (-1)
 
    FIX: original had this backwards (was shorting in accommodative regimes).
    FIX: column renamed to 'Econ_Strategy' to avoid collision with final 'Strategy'.
    """
    df['Econ_Strategy'] = np.where(df['Inflation'] > df['FedFunds'], 1, -1)
    return df
 
 
def TA_strategy(df):
    """
    MACD crossover signal:
      MACD > Signal line → bullish momentum → LONG  (+1)
      MACD < Signal line → bearish momentum → SHORT (-1)
 
    FIX: was (MACD > 1) — hardcoded dollar value, always true for large-cap prices.
    FIX: removed leading space from column name ' TA_Strategy'.
    """
    df['TA_Strategy'] = np.where(df['MACD'] > df['Signal'], 1, -1)
    return df
 
 
def add_full_strategy(df):
    """
    Combine both signals. Both must agree, else flat (0).
      +2 → both long  → LONG
      -2 → both short → SHORT
       0 → disagreement → FLAT
 
    Shift by 1 bar to prevent lookahead bias.
    """
    combined = df['Econ_Strategy'] + df['TA_Strategy']
    df['Strategy'] = np.where(combined == 2,  1,
                     np.where(combined == -2, -1, 0))
    df['Strategy'] = df['Strategy'].shift(1)   # execute on next bar's open
    return df
 
 
# ─── BACKTEST ─────────────────────────────────────────────────────────────────
 
def test_strategy(df):
    df['Strategy_Return'] = (
        1 + df['Close'].pct_change() * df['Strategy']
    ).cumprod() - 1
 
    # ── Performance metrics ───────────────────────────────────────────────────
    daily_ret = df['Close'].pct_change() * df['Strategy']
    daily_ret = daily_ret.dropna()
 
    trading_days = daily_ret[daily_ret != 0]
    sharpe = (trading_days.mean() / trading_days.std() * np.sqrt(252)
              if trading_days.std() > 0 else 0)
 
    # Max drawdown on strategy equity curve
    equity = (1 + daily_ret).cumprod()
    peak   = equity.cummax()
    dd     = (equity - peak) / peak
    max_dd = dd.min()
 
    # Win rate
    wins    = (trading_days > 0).sum()
    losses  = (trading_days < 0).sum()
    total   = wins + losses
    win_pct = wins / total * 100 if total > 0 else 0
 
    print(f"\n{'='*45}")
    print(f"  {TICKER} Strategy Performance")
    print(f"{'='*45}")
    print(f"  Asset Total Return  : {df['Asset_Return'].iloc[-1]*100:+.2f}%")
    print(f"  Strategy Total Return: {df['Strategy_Return'].iloc[-1]*100:+.2f}%")
    print(f"  Sharpe Ratio        : {sharpe:.3f}")
    print(f"  Max Drawdown        : {max_dd*100:.2f}%")
    print(f"  Win Rate            : {win_pct:.1f}%  ({wins}W / {losses}L)")
    print(f"  Days in Market      : {(df['Strategy'] != 0).sum()} / {len(df)}")
    print(f"{'='*45}\n")
 
    # ── Signal distribution ───────────────────────────────────────────────────
    sig_counts = df['Strategy'].value_counts().sort_index()
    print("  Signal distribution:")
    labels = {-1: 'SHORT', 0: 'FLAT', 1: 'LONG'}
    for k, v in sig_counts.items():
        print(f"    {labels.get(k, k):6s}: {v:5d} days  ({v/len(df)*100:.1f}%)")
    print()
 
    # ── Equity curve plot ─────────────────────────────────────────────────────
    plt.figure(figsize=(12, 5))
    plt.plot(df['Asset_Return']   * 100, label='Buy & Hold', color='grey', alpha=0.7)
    plt.plot(df['Strategy_Return'] * 100, label='Strategy',  color='blue')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.fill_between(df.index, df['Strategy_Return'] * 100, 0,
                     where=df['Strategy_Return'] >= 0, alpha=0.1, color='green')
    plt.fill_between(df.index, df['Strategy_Return'] * 100, 0,
                     where=df['Strategy_Return'] < 0, alpha=0.1, color='red')
    plt.title(f'{TICKER} — Cumulative Returns: Strategy vs Buy & Hold')
    plt.ylabel('Return (%)')
    plt.legend()
    plt.tight_layout()
    plt.show()
 
    # ── Drawdown plot ─────────────────────────────────────────────────────────
    plt.figure(figsize=(12, 3))
    plt.fill_between(dd.index, dd * 100, 0, color='red', alpha=0.4)
    plt.title('Strategy Drawdown')
    plt.ylabel('Drawdown (%)')
    plt.tight_layout()
    plt.show()
 
    return df
 
 
# ─── MAIN ─────────────────────────────────────────────────────────────────────
 
def main():
    df = get_data()
    df = add_interest_rate(df)
    df = add_inflation(df)
    df = add_bollinger_bands(df)
    df = economic_strategy(df)     # → 'Econ_Strategy' column
    df = TA_strategy(df)           # → 'TA_Strategy' column
    df = add_full_strategy(df)     # → 'Strategy' column (combines both)  FIX: was never called
    df = test_strategy(df)
    return df
 
 
if __name__ == '__main__':
    try:
        df = main()
    except Exception as e:
        print(f"Error: {e}")
    raise
 
# ─── Jupyter usage ────────────────────────────────────────────────────────────
# df = main()
# df = df.dropna()
# df.tail(20)
# df[['Close','MACD','Signal','Econ_Strategy','TA_Strategy','Strategy']].tail(30)
 
```

```python

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fredapi import Fred

# ─── CONFIG ───────────────────────────────────────────────────────────────────

TICKER   = 'TLT' 
INTERVAL = '1d'
PERIOD   = '1y' if INTERVAL == '1d' else '780d'

BB_WINDOW = 20
BB_STD    = 2.0
LOOKBACK  = 10000

FRED_KEY = '0673eba354376e034551ab39974ca711'
fred     = Fred(api_key=FRED_KEY)

# ─── DATA ─────────────────────────────────────────────────────────────────────

def get_data(ticker=TICKER, lookback=LOOKBACK):
    df = yf.download(ticker, interval=INTERVAL, period=PERIOD, progress=False)
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df['Asset_Return'] = (1 + df['Close'].pct_change()).cumprod() - 1
    for c in df.columns:
        df[f'{c}_change'] = df[c].pct_change() * 100

    subset = df.iloc[-lookback:].copy()

    plt.figure(figsize=(12, 4))
    plt.plot(subset['Close'])
    plt.title(f'{TICKER} Price Over Time')
    plt.tight_layout()
    plt.show()

    return subset.dropna()


def add_interest_rate(df):
    fedfunds = pd.DataFrame(fred.get_series('DFF'), columns=['FedFunds'])
    return df.join(fedfunds, how='left').ffill()


def add_inflation(df):
    fedinflation = pd.DataFrame(
        fred.get_series('CPIAUCSL', units='pc1'), columns=['Inflation']
    )
    return df.join(fedinflation, how='left').ffill()


# ─── INDICATORS ───────────────────────────────────────────────────────────────

def add_bollinger_bands(df, window=BB_WINDOW, num_std=BB_STD):
    df['BB_Mid']   = df['Close'].rolling(window).mean()
    rolling_std    = df['Close'].rolling(window).std()
    df['BB_Upper'] = df['BB_Mid'] + num_std * rolling_std
    df['BB_Lower'] = df['BB_Mid'] - num_std * rolling_std

    # %B: 0 = at lower band, 0.5 = middle, 1 = at upper band
    df['BB_PctB']  = (df['Close'] - df['BB_Lower']) / (df['BB_Upper'] - df['BB_Lower'])

    # Bandwidth: low = squeeze, big move coming
    df['BB_Width'] = (df['BB_Upper'] - df['BB_Lower']) / df['BB_Mid'] * 100

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6), sharex=True,
                                    gridspec_kw={'height_ratios': [3, 1]})

    ax1.plot(df['Close'],    color='black', linewidth=1.2, label='Price')
    ax1.plot(df['BB_Mid'],   color='blue',  linewidth=1.0, linestyle='--', label=f'SMA {window}')
    ax1.plot(df['BB_Upper'], color='red',   linewidth=0.8, label=f'Upper (+{num_std}σ)')
    ax1.plot(df['BB_Lower'], color='green', linewidth=0.8, label=f'Lower (-{num_std}σ)')
    ax1.fill_between(df.index, df['BB_Upper'], df['BB_Lower'], alpha=0.08, color='blue')
    ax1.set_title(f'{TICKER} Price + Bollinger Bands')
    ax1.legend(fontsize=8)

    ax2.plot(df['BB_PctB'], color='purple', linewidth=0.9)
    ax2.axhline(1.0, color='red',   linestyle='--', linewidth=0.8)
    ax2.axhline(0.5, color='grey',  linestyle=':',  linewidth=0.8)
    ax2.axhline(0.0, color='green', linestyle='--', linewidth=0.8)
    ax2.fill_between(df.index, df['BB_PctB'], 1.0,
                     where=(df['BB_PctB'] >= 1.0), alpha=0.2, color='red')
    ax2.fill_between(df.index, df['BB_PctB'], 0.0,
                     where=(df['BB_PctB'] <= 0.0), alpha=0.2, color='green')
    ax2.set_ylabel('%B')

    plt.tight_layout()
    plt.show()

    return df


# ─── STRATEGY SIGNALS ─────────────────────────────────────────────────────────

def economic_strategy(df):
    df['Econ_Strategy'] = np.where(df['Inflation'] > df['FedFunds'], 1, -1)
    return df


def TA_strategy(df):
    # %B < 0  → below lower band → oversold  → LONG  (+1)
    # %B > 1  → above upper band → overbought → SHORT (-1)
    # inside bands → FLAT (0)
    df['TA_Strategy'] = np.where(df['BB_PctB'] < 0.0,  1,
                        np.where(df['BB_PctB'] > 1.0, -1,
                                                        0))
    return df


def add_full_strategy(df):
    combined = df['Econ_Strategy'] + df['TA_Strategy']
    df['Strategy'] = np.where(combined == 2,  1,
                     np.where(combined == -2, -1, 0))
    df['Strategy'] = df['Strategy'].shift(1)
    return df


# ─── BACKTEST ─────────────────────────────────────────────────────────────────

def test_strategy(df):
    df['Strategy_Return'] = (
        1 + df['Close'].pct_change() * df['Strategy']
    ).cumprod() - 1

    daily_ret    = df['Close'].pct_change() * df['Strategy']
    daily_ret    = daily_ret.dropna()
    trading_days = daily_ret[daily_ret != 0]

    sharpe = (trading_days.mean() / trading_days.std() * np.sqrt(252)
              if trading_days.std() > 0 else 0)

    equity = (1 + daily_ret).cumprod()
    peak   = equity.cummax()
    dd     = (equity - peak) / peak
    max_dd = dd.min()

    wins    = (trading_days > 0).sum()
    losses  = (trading_days < 0).sum()
    total   = wins + losses
    win_pct = wins / total * 100 if total > 0 else 0

    print(f"\n{'='*45}")
    print(f"  {TICKER} Strategy Performance")
    print(f"{'='*45}")
    print(f"  Asset Total Return   : {df['Asset_Return'].iloc[-1]*100:+.2f}%")
    print(f"  Strategy Total Return: {df['Strategy_Return'].iloc[-1]*100:+.2f}%")
    print(f"  Sharpe Ratio         : {sharpe:.3f}")
    print(f"  Max Drawdown         : {max_dd*100:.2f}%")
    print(f"  Win Rate             : {win_pct:.1f}%  ({wins}W / {losses}L)")
    print(f"  Days in Market       : {(df['Strategy'] != 0).sum()} / {len(df)}")
    print(f"{'='*45}\n")

    sig_counts = df['Strategy'].value_counts().sort_index()
    print("  Signal distribution:")
    labels = {-1: 'SHORT', 0: 'FLAT', 1: 'LONG'}
    for k, v in sig_counts.items():
        print(f"    {labels.get(k, k):6s}: {v:5d} days  ({v/len(df)*100:.1f}%)")
    print()

    plt.figure(figsize=(12, 5))
    plt.plot(df['Asset_Return']    * 100, label='Buy & Hold', color='grey', alpha=0.7)
    plt.plot(df['Strategy_Return'] * 100, label='Strategy',   color='blue')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.fill_between(df.index, df['Strategy_Return'] * 100, 0,
                     where=df['Strategy_Return'] >= 0, alpha=0.1, color='green')
    plt.fill_between(df.index, df['Strategy_Return'] * 100, 0,
                     where=df['Strategy_Return'] < 0,  alpha=0.1, color='red')
    plt.title(f'{TICKER} — Cumulative Returns: Strategy vs Buy & Hold')
    plt.ylabel('Return (%)')
    plt.legend()
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 3))
    plt.fill_between(dd.index, dd * 100, 0, color='red', alpha=0.4)
    plt.title('Strategy Drawdown')
    plt.ylabel('Drawdown (%)')
    plt.tight_layout()
    plt.show()

    return df


# ─── MAIN ─────────────────────────────────────────────────────────────────────

def main():
    df = get_data()
    df = add_interest_rate(df)
    df = add_inflation(df)
    df = add_bollinger_bands(df)
    df = economic_strategy(df)
    df = TA_strategy(df)
    df = add_full_strategy(df)
    df = test_strategy(df)
    return df


try:
    df = main()
except Exception as e:
    print(f"Error: {e}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns

plt.rcParams['figure.figsize'] = (12, 8)
sns.set()

# ── Data ───────────────────────────────────────────────────
TSLA = yf.download('GBPEUR=X', period='5y', auto_adjust=True)
if isinstance(TSLA.columns, pd.MultiIndex):
    TSLA.columns = TSLA.columns.get_level_values(0)

MA = 21
TSLA['returns'] = np.log(TSLA['Close']).diff()
TSLA['MA']      = TSLA['Close'].rolling(MA).mean()
TSLA['ratio']   = TSLA['Close'] / TSLA['MA']
TSLA.dropna(inplace=True)

# ── Rolling percentiles — no lookahead ─────────────────────
TSLA['p95'] = TSLA['ratio'].rolling(252).quantile(0.95)
TSLA['p05'] = TSLA['ratio'].rolling(252).quantile(0.05)
TSLA.dropna(inplace=True)

# ── Signals ────────────────────────────────────────────────
TSLA['position'] = np.where(TSLA['ratio'] > TSLA['p95'], -1, np.nan)
TSLA['position'] = np.where(TSLA['ratio'] < TSLA['p05'],  1, TSLA['position'])
TSLA['position'] = TSLA['position'].ffill()

# ── Returns with cost ──────────────────────────────────────
TSLA['cost']         = TSLA['position'].diff().abs() * 0.001
TSLA['strat_return'] = TSLA['returns'] * TSLA['position'].shift() - TSLA['cost']

# ── Performance metrics ────────────────────────────────────
bh_ret    = TSLA['returns'].dropna()
st_ret    = TSLA['strat_return'].dropna()

bh_sharpe = bh_ret.mean() / bh_ret.std() * np.sqrt(252)
st_sharpe = st_ret.mean() / st_ret.std() * np.sqrt(252)
bh_total  = np.exp(bh_ret.sum()) - 1
st_total  = np.exp(st_ret.sum()) - 1

print(f"{'':20} {'Buy & Hold':>12} {'Strategy':>12}")
print(f"{'Total Return':20} {bh_total*100:>11.1f}% {st_total*100:>11.1f}%")
print(f"{'Sharpe Ratio':20} {bh_sharpe:>12.2f} {st_sharpe:>12.2f}")

# ── Plot ───────────────────────────────────────────────────
fig, axes = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

# Ratio + percentile bands
axes[0].plot(TSLA['ratio'],  color='purple', linewidth=0.8, label='Price/MA ratio')
axes[0].plot(TSLA['p95'],    color='red',    linestyle='--', linewidth=0.8, label='95th pct')
axes[0].plot(TSLA['p05'],    color='green',  linestyle='--', linewidth=0.8, label='5th pct')
axes[0].axhline(1.0, color='grey', linestyle=':', linewidth=0.6)
axes[0].set_title('TSLA Price / MA Ratio with Entry Bands')
axes[0].legend(fontsize=8)

# Position
axes[1].plot(TSLA['position'], color='steelblue', linewidth=0.8)
axes[1].axhline(0, color='grey', linewidth=0.5)
axes[1].set_title('Position  (+1 = Long, -1 = Short)')
axes[1].set_ylim(-1.5, 1.5)

# Cumulative returns
axes[2].plot(np.exp(bh_ret.cumsum()), label=f'Buy & Hold  ({bh_total*100:+.1f}%)', color='gold')
axes[2].plot(np.exp(st_ret.cumsum()), label=f'Strategy    ({st_total*100:+.1f}%)', color='green')
axes[2].axhline(1.0, color='grey', linestyle='--', linewidth=0.6)
axes[2].set_title('Cumulative Returns (after costs)')
axes[2].legend(fontsize=9)

plt.tight_layout()
plt.show()
```
